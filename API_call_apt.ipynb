```python
apt_codes = pd.DataFrame()
for city_code in unique_city_codes:
        params2 = {
            'serviceKey': key,
            'type': 'json',
            'cityCode': city_code, 
            'page': '1',
            'rowNum': '2000'
            }
        response2 = requests.get(url2, params=params2)
        if response2.status_code == 200:
            try:
                json_obj2 = json.loads(response2.text)
                select2 = json_obj2['data']['list']
                dta2 = pd.DataFrame.from_dict(select2)
                apt_codes = pd.concat([apt_codes, dta2], ignore_index=True)
            except json.JSONDecodeError:
                # Handle invalid JSON response here if needed
                print(f"Invalid JSON response for city_code: {city_code}")
        else:
            # Handle non-successful response here if needed
            print(f"Request failed for city_code: {city_code}")
```
